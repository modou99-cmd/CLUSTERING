```python
# Conversion pandas -> Spark
df_spark = spark.createDataFrame(df)

from pyspark.ml.feature import VectorAssembler

# CORRECTION : convertir Index en liste Python
input_cols = list(df.columns[:-1])  # <= FIX

assembler = VectorAssembler(
    inputCols=input_cols,
    outputCol="features"
)

df_spark_vec = assembler.transform(df_spark)
data_spark = df_spark_vec.select("features")

data_spark.show(5)
```

```python
# K-means Scikit-learn
from sklearn.cluster import KMeans
import time
X = df.drop('target',axis=1).values
t0=time.time()
km = KMeans(n_clusters=5,n_init=10)
labels = km.fit_predict(X)
t_sklearn = time.time()-t0
t_sklearn
```

```python
# K-means Spark
from pyspark.ml.clustering import KMeans as KMS
t0=time.time()
kms = KMS(k=5, seed=42)
model_spark = kms.fit(data_spark)
pred_spark = model_spark.transform(data_spark)
t_spark = time.time()-t0
t_spark
```

```python
# Évaluation silhouette
from sklearn.metrics import silhouette_score
sil_sklearn = silhouette_score(X, labels)
from pyspark.ml.evaluation import ClusteringEvaluator
evaluator = ClusteringEvaluator(featuresCol='features')
sil_spark = evaluator.evaluate(pred_spark)
sil_sklearn, sil_spark
```
